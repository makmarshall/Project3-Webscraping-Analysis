## Read in the webscraped datafiles

```python
def read_in_csv(csv):
    datafile = csv
    return pd.read_csv(datafile)
```

```python
Worldwide2023 = read_in_csv('Worldwide2023.csv')
DomesticDaily2023 = read_in_csv('DomesticDaily2023.csv')
DomesticWeekly2023 = read_in_csv('DomesticWeekly2023.csv')
DomesticMonthly2023 = read_in_csv('DomesticMonthly2023.csv')
DomesticYearly = read_in_csv('DomesticYearly.csv')
```

```python
# replace the whole dataframe with integers for the money instead.
def int_for_df(dataframe, col):
    for i in range(0, len(dataframe)):
        c = dataframe[col][i]
        c = c.lstrip('$')
        c = c.replace(',', '')
        c = int(c)
        dataframe[col][i] = c
    return dataframe
```
